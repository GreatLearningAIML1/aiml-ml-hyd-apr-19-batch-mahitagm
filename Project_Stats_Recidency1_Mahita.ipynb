```python
import pandas as pd
import numpy as np

fb = pd.read_csv('Laliga.csv')
```

```python
fb.head()
```

```python
fb.tail()
```

```python
fb.dtypes
```

```python
fb['Pos']=fb['Pos'].replace('-', 0)
fb['Seasons']=fb['Seasons'].replace('-', 0)
fb['Points']=fb['Points'].replace('-', 0)
fb['GamesPlayed']=fb['GamesPlayed'].replace('-', 0)
fb['GamesWon']=fb['GamesWon'].replace('-', 0)
fb['GamesDrawn']=fb['GamesDrawn'].replace('-', 0)
fb['GamesLost']=fb['GamesLost'].replace('-', 0)
fb['GoalsFor']=fb['GoalsFor'].replace('-', 0)
fb['GoalsAgainst']=fb['GoalsAgainst'].replace('-', 0)
fb['Champion']=fb['Champion'].replace('-', 0)
fb['Runner-up']=fb['Runner-up'].replace('-', 0)
fb['Third']=fb['Third'].replace('-', 0)
fb['Fourth']=fb['Fourth'].replace('-', 0)
fb['Fifth']=fb['Fifth'].replace('-', 0)
fb['Sixth']=fb['Sixth'].replace('-', 0)
fb['T']=fb['T'].replace('-', 0)
fb['Debut']=fb['Debut'].replace('-', 0)
fb['Since/LastApp']=fb['Since/LastApp'].replace('-', 0)
fb['BestPosition']=fb['BestPosition'].replace('-', 0)

```

```python
fb.tail()
```

```python
fb['Debut1']= fb['Debut'].str.slice(0,4)
fb['Debut2']=fb['Debut'].str.slice(0,2)+fb['Debut'].str.slice(5,7)
```

```python
fb.tail()
```

```python
fb.dtypes
```

```python
fb['Debut1'] = pd.to_numeric(fb['Debut1'])
fb['Debut2']=pd.to_numeric(fb['Debut2'])
fb['Points']=pd.to_numeric(fb['Points'])
fb['GoalsFor']=pd.to_numeric(fb['GoalsFor'])
fb['GoalsAgainst']=pd.to_numeric(fb['GoalsAgainst'])
fb['GamesWon']=pd.to_numeric(fb['GamesWon'])
fb['GamesPlayed']=pd.to_numeric(fb['GamesPlayed'])
```

```python
fb.dtypes
```

```python
fb[(fb['Debut1'] >= 1930) & (fb['Debut1'] <= 1980 ) & (fb['Debut2'] >= 1930) & (fb['Debut2'] <= 1980 )].Team
```

```python
fb['Rank_points'] = fb['Points'].rank()
```

```python
fb.head()
```

```python
fb[(fb['Rank_points'] >= 1) & (fb['Rank_points'] <= 5 )].Team
```

```python
def Goal_diff_count(GoalsFor,GoalsAgainst) :
    return GoalsFor-GoalsAgainst
```

```python
fb['Goal_diff_count'] = Goal_diff_count(fb['GoalsFor'],fb['GoalsAgainst'])
```

```python
fb[fb['Goal_diff_count'] == fb.Goal_diff_count.max()].Team
```

```python
fb[fb['Goal_diff_count'] == fb.Goal_diff_count.min()].Team
```

```python
fb['Winning_Percentage'] = (fb['GamesWon']/fb['GamesPlayed'])*100
```

```python
fb['Winning_Percentage'] 
```

```python
fb['Winning_Percentage'].fillna(0,inplace = True)
```
